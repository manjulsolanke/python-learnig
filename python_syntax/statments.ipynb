### Python Indentation 

```python
if 3 < 5:
    print('3 is less than 5') # Proper indentation, with four spaces
if 10 > 20:
 print('10 is greater than 20')   # Even though indentation is not proper, it will work
if 7 == 7:
    print('7 is equal to 7')  # Proper indentation, with four spaces
 print('This line has inconsistent indentation')  # This will cause an IndentationError
```
