### Unpack a Collection
If you have a collection of values in a list, tuple etc. Python allows you to extract the values into variables. This is called unpacking.

```python

fruit = "apple", "cherry", "banana"  # Tuple unpacking
z, y, x  = fruit # Unpacking the tuple into variables
print(x) # Outputs: banana
print(y) # Outputs: cherry
print(z) # Outputs: apple
```

```python
x = y = z = "Hello World"
print(x)
print(y)
print(z)
```

```python
def calculate(num1, num2):
    sum_result = num1 + num2
    sub_result = num1 - num2
    mul_result = num1 * num2 
    div_result = num1 / num2
    return  sum_result, sub_result, mul_result, div_result
sum_result, sub_result, mul_result, div_result = calculate(50, 10)
print("Sum:", sum_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Division:", div_result)
```

### Python - Global Variables
Variables that are created outside of a function, are known as global variables. Global variables can be used by everyone, both inside of functions and outside.



```python
x = "awesome" # global variable
print("Python is " + x)
def myfunc():
    x = "fantastic" # local variable
    print("Python is " + x)

myfunc()
```

Global Keyword
Normally, when you create a variable inside a function, that variable is local, and can only be used inside that function.
To create a global variable inside a function, you can use the global keyword.



```python
x = "awesome" # global variable
def myfunc():
    global x # use the global keyword to modify the global variable
    y  = "fantastic" # now x is global due to the global keyword
    print("Python is " + x) # Accessing the global variable x
    print("Python is " + y) # Accessing the local variable y

myfunc()   
```

```python
x = 'awesome'
def myfunc():
  global x
  x = 'fantastic' # now x is changed to 'fantastic', because of the global keyword that modifies the global variable. It does not create a local variable.
myfunc()
print('Python is ' + x)
```
