A python program that displays stars in right angled trianguler form using a single loop.


```python
for i in range(1, 11):
    print('* '*(i))
```

A python program to display the stars in an equilateral triangular form using a single for loop

```python
n=40
for i in range(1, 11):
    print(' '*n, end='')
    print('* '*(i))
    n-=1
```

A python program that displays stars in right angled triangular form using nested loop

```python
for i in range(1, 11): # Outer loop: controls the row ( 1 to 10)
    for j in range(1, i+1): # inner loop: controls the star per row
        print('*', end='') # Print the stars on the same line
    print()
```
